# Data preparation

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Load original data
data = pd.read_excel('/home/mikhail91/Desktop/popularity-728days_my.xls')
data.columns
```

```python
#Select data
selection = ((data['Now'] - data['Creation-week']) > 26)&((data['Now'] - data['FirstUsage']) > 26)&((data[78] - data[1]) != 0)
data_sel = data[selection].copy()
```

```python
data_b = data_sel.copy()
periods = range(1,104)
#Get week's usages
for i in periods:
    if i!=1:
        data_b[i] = data_sel[i] - data_sel[i-1]
```

```python
for i in range(0, 200):
    cuv = data_b[periods].irow(i).values
    plt.subplot(1,1,1)
    plt.bar(periods, cuv, width=1, bottom=0)
    plt.xlim(1,104)
    plt.xlabel('Weeks')
    plt.ylabel('Nb of usages')
    plt.show()
```
