# About

I take 1-78 weeks and make forecast for the last 26 weeks (0 or 1).
Forecast is 1 when last N weeks the file was not been used. 
Forecast is 0 in the other case.

I make pre-selection of the data. Look attantively!!!

Look on the results!!!

# Data preparation

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Load original data
data = pd.read_excel('/home/mikhail91/Desktop/popularity-728days_my.xls')
data.columns
```

```python
#Select data
selection = ((data['Now'] - data['Creation-week']) > 26)&((data['Now'] - data['FirstUsage']) > 26)#&((data[78] - data[1]) != 0)
data_sel = data[selection].copy()
```

# Make predictions

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
```

```python
#Get true labels
y_true = ((data_sel[104] - data_sel[78]) == 0).values*1

#Prediction for N last zero weeks
def SimplePrediction(N):
    
    y_predict = ((data_sel[78] - data_sel[74-N]) == 0).values*1
    
    accuracy = accuracy_score(y_true, y_predict)
    f1 = f1_score(y_true, y_predict)
    precision = precision_score(y_true, y_predict)
    
    released_memory = (data_sel['DiskSize'].values*y_predict).sum()
    good_memory = (data_sel['DiskSize'].values*y_predict*y_true).sum()
    part_memory = float(good_memory)/float(released_memory)*100
    
    return accuracy, f1, precision, released_memory, good_memory, part_memory

#Predcitions for severat last zero weeks
def PredictionCurves(wmin=1, wmax=50):
    x = range(wmin, wmax+1)
    
    arr_accuracy = []
    arr_f1 = []
    arr_precision = []
    
    arr_released = []
    arr_good = []
    arr_part = []
    
    #Predict
    for i in range(wmin, wmax+1):
        accuracy, f1, precision, released_memory, good_memory, part_memory = SimplePrediction(i)
        arr_accuracy.append(accuracy)
        arr_f1.append(f1)
        arr_precision.append(precision)
        #Memory
        arr_released.append(released_memory)
        arr_good.append(good_memory)
        arr_part.append(part_memory)
        
    #Plotting results
    #Metrics
    plt.subplot(1,1,1)
    plt.plot(x, arr_accuracy, 'r', label = 'accuracy')
    plt.plot(x, arr_f1, 'b', label = 'f1')
    plt.plot(x, arr_precision, 'g', label = 'precision')
    plt.title('Prediction Results')
    plt.legend(loc = 'best')
    plt.show()
    #Memory
    #Print Info
    print 'Total Memory = ', (data_sel['DiskSize'].values).sum()
    print 'Memory Can Be Released = ', (data_sel['DiskSize'].values*y_true).sum()
    #Plotting
    plt.subplot(1,1,1)
    plt.plot(x, arr_released, 'b', label = 'released memory')
    plt.plot(x, arr_good, 'r', label = 'good memory')
    plt.title('Memory')
    plt.legend(loc = 'best')
    plt.show()
    #Part Memory
    plt.subplot(1,1,1)
    plt.plot(x, arr_part, 'g', label = 'part memory (%)')
    plt.title('Part Memory')
    plt.legend(loc = 'best')
    plt.show()
    
```

```python
PredictionCurves(0,73)
```
