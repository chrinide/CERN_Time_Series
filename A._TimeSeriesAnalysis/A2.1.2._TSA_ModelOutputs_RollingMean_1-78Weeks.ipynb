# Some usefull methods

```python
#Holt's lineral model with Grid Search optimization
from pycast.optimization import GridSearch
from pycast.errors import SymmetricMeanAbsolutePercentageError as SMAPE
from pycast.errors.meansquarederror import MeanSquaredError as MSE
from pycast.common.timeseries import TimeSeries
from pycast.common.json_encoder import PycastEncoder
from pycast.methods.exponentialsmoothing import HoltMethod
from pycast.methods.exponentialsmoothing import HoltWintersMethod
from pycast.common.timeseries import TimeSeries

def MyESOptima(serie, forecast = 52):
    array_2d = np.array([range(0, serie.shape[0]), serie], ndmin=2,).transpose()
    t_serie = TimeSeries.from_twodim_list(datalist = array_2d)
    
    hwm = HoltMethod( valuesToForecast = forecast)
    gridSearch = GridSearch( MSE, precision= -2)
    optimal_forecasting, error, optimal_params = gridSearch.optimize(t_serie, [hwm])
    
    t_res = optimal_forecasting.execute(t_serie)
    res = np.array(t_res.to_twodim_list())[:,1]
    
    y_0 = res[-forecast]
    y_1 = res[-(forecast-1)]
    y_f = res[-1]
    slope = float(y_f - y_0)/float(forecast)
    
    return res, slope, y_0, y_1, y_f, optimal_forecasting, error, optimal_params

def MyES(serie, alpha=0.99, gamma=0.35, forecast=52):
    array_2d = np.array([range(0, serie.shape[0]), serie], ndmin=2,).transpose()
    t_serie = TimeSeries.from_twodim_list(datalist = array_2d)
    
    method = HoltMethod(smoothingFactor=alpha, trendSmoothingFactor=gamma, valuesToForecast=forecast)
    t_res = method.execute(t_serie)
    res = np.array(t_res.to_twodim_list())[:,1]
    
    y_0 = res[-forecast]
    y_1 = res[-(forecast-1)]
    y_f = res[-1]
    slope = float(y_f - y_0)/float(forecast)
    
    return res, slope, y_0, y_1, y_f

def MyESSimp(optimal_forecasting, serie, forecast = 52):
    array_2d = np.array([range(0, serie.shape[0]), serie], ndmin=2,).transpose()
    t_serie = TimeSeries.from_twodim_list(datalist = array_2d)
    
    t_res = optimal_forecasting.execute(t_serie)
    res = np.array(t_res.to_twodim_list())[:,1]
    
    y_0 = res[-forecast]
    y_1 = res[-(forecast-1)]
    y_f = res[-1]
    slope = float(y_f - y_0)/float(forecast)
    
    return res, slope, y_0, y_1, y_f
```

```python
#For getting residual vector
def Residuals(train, predict):
    errors = []
    for i in range(0, len(train)):
        if np.isnan(predict[i]) == False:
            errors.append(train[i] - predict[i])
    
    return np.array(errors)

#Compute SSE
def SSE(errors):
    sse = 0
    for i in errors:
        sse = sse + i*i
        
    return sse

#Compute AIC
def AIC(N, sse, k):
    return N*np.log(sse/N) + 2*(k+2)
```

```python
#For getting probability of zero or less value
def Get_Negative_Proba(time_serie, optimal_forecasting , N = 200, forecast = 52):
    
    prediction, slope, y_0, y_1, y_f = MyESSimp(optimal_forecasting, time_serie, forecast = forecast)
    residuals = Residuals(time_serie, prediction)
    
    mean_value = time_serie.mean()
    predict_last_point = []
    slopes = []
    ys_0 = []
    ys_1 = []
    ys_f = []
    
    #Get statisitc of the prediction
    for i in range(0,N):
        random_value = np.random.normal(0, residuals.std()+0.001, len(time_serie))
        #random_value = np.random.poisson(mean_value, len(time_serie)) - mean_value
        time_and_random = time_serie + random_value
        prediction, slope, y_0, y_1, y_f = MyESSimp(optimal_forecasting, time_and_random, forecast = forecast)
        #plt.subplot(10,2,1)
        plt.plot(prediction, color = '0.9')
        #predict_last_point.append(prediction[-1:])
        predict_last_point.append(slope)
        slopes.append(slope)
        ys_0.append(y_0)
        ys_1.append(y_1)
        ys_f.append(y_f)
        
    slopes = np.array(slopes)
    ys_0 = np.array(ys_0)
    ys_1 = np.array(ys_1)
    ys_f = np.array(ys_f)
        
    #Get probability of the zero or negative value of the prediction
    #hist, binns = np.histogram(predict_last_point, bins = [np.min(p), 0, np.max(p)])
    hist, binns = np.histogram(predict_last_point, bins = [-np.inf, 0, np.inf])
    probability_of_negative = float(hist[0])/N
    #plt.show()
    
    return probability_of_negative ,np.array(predict_last_point), slopes.std(), ys_0.std(), ys_1.std(), ys_f.std()
```

```python
# All for predicting one file
def Predict_For_One(time_serie1_rolling_mean, forecast=26):
    
    #last_periods = periods_rolling_mean
    
    #Make model evaluation
    prediction, slope, y_0, y_1, y_f, optimal_forecasting, error, optimal_params = MyESOptima(time_serie1_rolling_mean, forecast=forecast)
    
    #Get prediction error
    neg_prob, predicts, slope_std, y_0_std, y_1_std, y_f_std = Get_Negative_Proba(time_serie1_rolling_mean, optimal_forecasting = optimal_forecasting , N=200, forecast = forecast)
    
    return neg_prob, time_serie1_rolling_mean, prediction, predicts, slope, y_0, y_1, y_f, slope_std, y_0_std, y_1_std, y_f_std
    
```
